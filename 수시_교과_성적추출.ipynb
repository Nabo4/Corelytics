```python
import os
import re
import pandas as pd

path = r"data\수시_컴과_학종외_생기부_txt\수시_교과_txt"

subjects = {
    "국어": ["국어", "문학", "화작", "언어"],
    "영어": ["영어"],
    "수학": ["수학", "미적분", "기하", "확률"],
    "사회": ["사회", "한국사", "윤리", "경제", "정치", "지리"],
    "과학": ["과학", "물리", "화학", "생명", "지구"]
}

files = [f for f in os.listdir(path) if f.endswith(".txt")]

rows = []

for file in files:
    with open(os.path.join(path, file), "r", encoding="utf-8") as f:
        lines = f.readlines()

    for line in lines:
        for subject, keywords in subjects.items():
            if any(k in line for k in keywords):
                grades = re.findall(r"[A-F]\(\d+\)", line)
                for g in grades:
                    rows.append({
                        "학생": file.replace(".txt", ""),
                        "과목": subject,
                        "등급": g[0]
                    })

df = pd.DataFrame(rows)

grade_map = {"A": 5, "B": 4, "C": 3, "D": 2, "E": 1, "F": 0}
df["점수"] = df["등급"].map(grade_map)

student_df = (
    df
    .groupby(["학생", "과목"])["점수"]
    .mean()
    .reset_index()
    .pivot(index="학생", columns="과목", values="점수")
)

student_df.to_excel("학생별_국영수사과_성적.xlsx")

```
